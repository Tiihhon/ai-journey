```python
# Создаём датасет вручную (если файл не скачался)
import pandas as pd
import numpy as np

np.random.seed(42)  # Для повторяемости
n_samples = 1000
area = np.random.uniform(30, 200, n_samples)
rooms = np.random.randint(1, 6, n_samples)
price = 100000 * area + 500000 * rooms + np.random.normal(0, 1000000, n_samples)

df = pd.DataFrame({
    'area': area,
    'rooms': rooms,
    'price': price
})

# Сохраняем в файл
df.to_csv('housing.csv', index=False)
print("✅ Файл `housing.csv` создан! Первые 5 строк:")
print(df.head())
```

```python
# Основной код (теперь файл есть!)
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# График
plt.scatter(df['area'], df['price'], color='blue', alpha=0.5)
plt.xlabel('Площадь (кв.м)')
plt.ylabel('Цена (руб)')
plt.title('Зависимость цены от площади')
plt.show()

# Модель
X = df[['area']].values
y = df['price'].values
model = LinearRegression()
model.fit(X, y)

print(f"Формула: price = {model.coef_[0]:.2f} * area + {model.intercept_:.2f}")

# Предсказание
area_new = 75
price_pred = model.predict([[area_new]])[0]
print(f"Дом 75 кв.м → цена ≈ {price_pred:,.0f} руб")
```
